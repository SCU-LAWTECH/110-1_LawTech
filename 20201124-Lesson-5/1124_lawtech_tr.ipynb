```python
driver = webdriver.Chrome("C:/Users/User/Downloads/chromedriver_win32/chromedriver")  
driver.get("https://www.lawplus.com.tw/#/searchReport?query=%25E8%25BB%258A%25E7%25A6%258D")
sleep(1)
b = []
for e in range(1,4):
    for i in range(1,11):
        while True:
            try:
                driver.find_element_by_xpath("//ol/li[{}]/h4/a".format(i)).click()
                sleep(2)
                d = {}
                d["ptzu_no"] = driver.find_element_by_xpath("//div[@class='reportInfo nine columns']/h5").text  #抓裁判字號文字
                d["ptzu_date"] = driver.find_element_by_xpath("//div[@class='reportInfo nine columns']/p").text  #抓裁判日期
                d["ptzu_content"] = driver.find_element_by_xpath("//div[@class='content']/pre[3]").text  #抓主文 
                b.append(d)
                driver.back()
                sleep(1)
                break
            except:
                print(e,i)
                pass
    driver.find_element_by_xpath("//a/i[@class='icon-right-open']").click() 
    sleep(3)
    
```
