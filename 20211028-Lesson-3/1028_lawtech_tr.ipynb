# if

```python
#if else
a = 10
if a == 10:
    print("Ten")
else:
    print("Not Ten")
#講縮排
```

```python
#elif
b = 16
if b > 15:
    print("Bigger")
elif b > 10:
    print("Equal")
else:
    print("Smaller")
#講elif順序
```

# for

```python
#for in list
for i in [1,"app;e",3,4]:
    print(i)
```

```python
#for in string
for e in "ABCD":
    print(e)
```

```python
#range
for f in range(1,11,2):
    print(f)
```

```python
#range預設 print end
for g in range(10):
    print(g,end="\\n")
```

```python
#巢狀迴圈
for h in range(0,10):
    print(h,end=" ")
    for j in ["A","B","C","D"]:
        print(j,end=" ")
    print("")
```

```python
#break
for q in range(81,100,2):
    if q == 87:
        break
    else:
        print(q)
```

```python
#coontinue
for q in range(81,100,2):
    if q == 87:
        print("cotinue1")
        continue
        print("continue2")
    else:
        print(q)
```

```python
#pass
for q in range(81,100,2):
    if q == 87:
        pass
    else:
        print(q)
```

```python
#練習
for i in range(1,10):
    for g in range(1,10):
        print(i,"X",g,"=",i*g)
```

```python
#炫技
print(' \n'.join([' \n'.join(["%s x %s = %s"%(i,j,i*j) for j in range(1,10)]) for i in range(1,10)]))
```

```python
#error處理try
c = input("請輸入成績")
gate = True
while gate:
    try:
        c = int(c)
        gate = False
    except:
        print("輸入整數!!!")
        c = input("請輸入成績")
if c > 100:
    print("超過一百分")
elif c >= 60:
    print("及格")
elif c >= 50:
    print("要補考")
else:
    print("暑修見")
#講while也有break
```

```python
#練習
name = ["湯德宗大法官","許宗力大法官","王澤鑑大法官"]
while True:
    user = input("輸入使用者名稱")
    if user in name:
        print("歡迎",user)
        break
    else:
        print("請重新輸入")
```

```python
#format
a = "123{}56{}".format(4,"七")
print(a)
print("123",4,"56","七")
```

```python
#%
#%s 是以字串輸出，%f 是以浮點數輸出、%d 是以十進位整數輸出
a = "123%s56%d"%("四",7)
print(a)
```

```python
#zip
a = [1,2,3]
b = ["1","2","3"]
c = ["one","two","three"]
d = zip(a,b,c)
print(type(d))
e = list(d)
print(e)
f = zip(e[0],e[1],e[2])
print(list(f))
```

```python
#enumerate
a = ["one","two","three"]
b = enumerate(a)
c = enumerate(a,start=1)
print(type(b))
print(list(b))
print(list(c))
for i in enumerate(a):
    print(i)
for e,f in enumerate(a):
    print(e,f)
```
